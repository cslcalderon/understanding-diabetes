```python
#IMPORTING MY DATA
import matplotlib.pyplot as plt
import numpy as np
import pandas as pand #will help us read .csv file

#Getting data from the file and reading it through pandas due to combination of numbers and words in the file 
data = (pand.read_csv('diabetes.csv'))

#Loading the data from the different columns using pandas 
glucoseDF = pand.read_csv('diabetes.csv', usecols= [2])
waistDF = pand.read_csv('diabetes.csv', usecols= [16])
hipDF = pand.read_csv('diabetes.csv', usecols= [17])

#The data was loaded into a data frame variable value, but they have to be converted to 
#lists to be able to be parsed through and manipulated accordingly. 
#These values load them into a 2D array though, and they will be convereted into 1D because
#they are just one column and it's inconvinient to parse through the format this way. 
waistMeasurement = waistDF.values.tolist()
hipMeaasurement = hipDF.values.tolist()
glucoseMesurement = glucoseDF.values.tolist()

#This is loading the data into a 1D array to make data parsing easier for glucoses readings
glucoseReadings =[]
for i in range(0, 403):
    glucoseReadings.append(glucoseMesurement[i][0])

#This is loading ratios into a 1D array of waist/hip measurments from the list values 
#loaded in before
ratios = []
for i in range(0, 403):
    ratios.append(waistMeasurement[i][0]/ hipMeaasurement[i][0])

#The below lines of code are making arrays of nondiabetic, prediabtic, and diabetic subjects from the study, as well 
#as making array of all the subjects and their diabetic status
subjects = []
noDibetes = []
preDiabetes = []
hasDiabetes = []

#This is creating 403 values in the subjects array, which is the total amount of subjects in the study
for i in (range(0, 403)):
    subjects.append(i)

#This is dividing diabetic subject's ratios into arrays and setting diabetic status for whole study population
for i in range(0, len(subjects)):
    if glucoseReadings[i] < 99: # below 99 is considered non-diabetic
        subjects[i] = 0 # setting status to non diabetic
        noDibetes.append(ratios[i]) #adding the subject's waist/hip ratio to non diabetic list
    elif glucoseReadings[i] < 125: # 99-125 glucosse level is prediabetic 
        subjects[i] = 1  # setting status to  prediabetic
        preDiabetes.append(ratios[i])#adding the subject's waist/hip ratio to prediabetic list
    else:
        subjects[i] = 2 # setting status to diabetic
        hasDiabetes.append(ratios[i]) #adding the subject's waist/hip ratio to diabetic list
        

# This was used to verify the loading of the data
print(ratios)
print(waistMeasurement)
print(hipMeaasurement)

rows = 403
columns = 19
 
```

```python
# Plotting a scatter plot between ratios vs. glucose measurements
ascat = plt.axes()
fontsize = 15

#Labeling the graph
ascat.set_xlabel('Waist/Hip Ratio (in)', fontsize= fontsize)
ascat.set_ylabel('Blood Sugar Level (mg/dL)', fontsize= fontsize)
ascat.set_title('Waist/Hip Ratio (in) vs. Blood Sugar Levels (mg/dL) for Random population of Americans')
ascat.scatter(ratios, glucoseReadings)

#adding description to the graph
ascat.text(1.2,200,"Description of graph: \nThis graph is graphing the possible relationship \nbetween waist/hip measurement ratios and \nblood sugar levels for a random group of African \nAmericans. \n\nBlood sugars above 99 mg/dL is considered prediabetic. \nBlood sugars above 120 mg/dL is considered diabetic", fontsize = 10)
ascat.spines['top'].set_visible(False) #getting rif of top and right axes lines 
ascat.spines['right'].set_visible(False)
```
